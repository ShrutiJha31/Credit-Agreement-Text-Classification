## Load Data

```python
data = pd.read_json(TRAINING_DATA_JSON_FILE)
data
```

```python
data_features = data.loc[:,data.columns != 'CATEGORY']  
data_features
```

```python
data_targets = data.loc[:,data.columns != 'STATEMENT']  
data_targets
```

## Creating a Vocabulary

```python
vectorizer = CountVectorizer(stop_words='english')
```

```python
all_features = vectorizer.fit_transform(data.STATEMENT)
vectorizer.vocabulary_
```

```python
x_train, x_test, y_train, y_test = train_test_split(
                        all_features, data.CATEGORY,
                test_size = 0.30, random_state = 42)

#---- Due to random splitting and small dataset issue might arise due to some labels are present in y_test but not in y_train 
#---- To address such issue I have evenly split the data without using sklearn kit in my Naive Bayes Model 
#---- Refer File Naive Bayes Step 1 - Pre-Processing Data
```

## Training and Testing the Model - Naive Bayes

```python
classifier = MultinomialNB()
classifier.fit(x_train, y_train)
```

```python
predictions = classifier.predict(x_test)
print(classification_report(y_test, predictions))
```

## Training and Testing the Model - SVM

```python
classifier_svm = SVC()
classifier_svm.fit(x_train, y_train)
```

```python
predictions = classifier_svm.predict(x_test)
print(classification_report(y_test, predictions))
```

## Training and Testing the Model - Decision Trees

```python
classifier_dt = DecisionTreeClassifier()
classifier_dt = classifier_dt.fit(x_train,y_train)
```

```python
predictions = classifier_dt.predict(x_test)
print(classification_report(y_test, predictions))
```
